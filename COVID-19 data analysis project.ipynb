```python
import pandas as pd

# Load the local dataset
df = pd.read_csv("owid-covid-data.csv")

# Preview the columns and first few rows
print("Columns:\n", df.columns)
print("\nFirst 5 rows:\n", df.head())

# Filter for selected countries
countries = ["Kenya", "India", "Congo", "Uganda"]
df_filtered = df[df["location"].isin(countries)]

# Select only necessary columns
columns_of_interest = ["location", "date", "total_cases", "total_deaths", "people_vaccinated"]
df_selected = df_filtered[columns_of_interest]

# Check for missing values
missing_values = df_selected.isnull().sum()
print("\nMissing values:\n", missing_values)

# Preview cleaned data
print("\nCleaned preview:\n", df_selected.head())
```
